# Data Practitioner Salary Analysis

This notebook sets up an API-first workflow in Python to answer:

**How much do we get paid?**

The analysis is designed to compare salary variation by:

- **role descriptor**
- **state**

It uses the **BLS OEWS API** as the primary source for salary data and keeps the role-to-occupation mapping explicit so the methodology is transparent and reproducible.

## 3. BLS OEWS series builder

The OEWS API uses a structured series ID. For this project, we will pull:

- **state-level** data
- **all industries**
- **occupation-specific** wages
- **annual mean wage** by default

Measure codes:

- `04` = annual mean wage
- `13` = annual median wage


```python
API_URL = "https://api.bls.gov/publicAPI/v2/timeseries/data/"
MEASURE_CODE = "04"  # change to "13" for annual median wage

def build_oews_series_id(state_area_code: str, soc_code: str, measure_code: str = "04") -> str:
    soc = soc_code.replace("-", "")
    # OE + U + S + area(7) + industry(000000) + occupation(6) + data_type(2)
    return f"OEUS{state_area_code}000000{soc}{measure_code}"

def chunked(iterable, size):
    iterable = iter(iterable)
    while True:
        batch = list(islice(iterable, size))
        if not batch:
            break
        yield batch
```

## 4. Build the request list

This creates one BLS series per `(role, state)` pair.


```python
series_ids = []
series_meta = {}

for role, soc in ROLE_TO_SOC.items():
    for state_abbr, state_area_code in STATE_FIPS.items():
        sid = build_oews_series_id(state_area_code, soc, MEASURE_CODE)
        series_ids.append(sid)
        series_meta[sid] = {
            "role": role,
            "soc_code": soc,
            "state": state_abbr,
        }

len(series_ids)
```

## 5. Pull data from the BLS API

The public BLS API accepts up to 50 series IDs per request, so we batch the calls.


```python
all_rows = []

for batch in chunked(series_ids, 50):
    payload = {"seriesid": batch}
    response = requests.post(API_URL, json=payload, timeout=30)
    response.raise_for_status()
    result = response.json()

    for series in result["Results"]["series"]:
        sid = series["seriesID"]
        points = series.get("data", [])
        if not points:
            continue

        latest = points[0]  # BLS returns latest period first
        meta = series_meta[sid]
        state = meta["state"]
        region = STATE_TO_REGION.get(state, "Unknown")

        all_rows.append({
            "role": meta["role"],
            "soc_code": meta["soc_code"],
            "state": state,
            "region": region,
            "work_location": region if region != "Unknown" else None,  # will be used for regional analysis
            "year": latest.get("year"),
            "period": latest.get("period"),
            "salary": pd.to_numeric(latest.get("value"), errors="coerce"),
            "series_id": sid,
        })

df = pd.DataFrame(all_rows)

# Add a placeholder level column - you should populate this with actual data if available
# For now, we'll divide by salary percentiles to create example levels
df['level'] = pd.cut(df['salary'], bins=3, labels=['Junior', 'Mid-Level', 'Senior'], duplicates='drop')

print(f"Data shape: {df.shape}")
print(f"Regions in dataset: {df['region'].unique()}")
print(f"Levels created: {df['level'].unique()}")
df.head(10)
```

## 7. Summaries for analysis

Create tables that support the final story.


```python
print("="*60)
print("SALARY SUMMARIES")
print("="*60)

# 1. Average salary by role (original)
salary_by_role = (
    df.groupby('role', as_index=False)['salary']
      .agg(['mean', 'median', 'count'])
      .sort_values('mean', ascending=False)
)
print("\n1. Average salary by role:")
print(salary_by_role)

# 2. New: Average salary by level and role
salary_by_level_role = (
    df.groupby(['level', 'role'], as_index=False)['salary']
      .agg(['mean', 'median', 'count'])
      .sort_values(['level', 'mean'], ascending=[True, False])
)
print("\n2. Average salary by level and role:")
print(salary_by_level_role)

# 3. New: Average salary by region (with Remote as distinct category)
salary_by_region = (
    df.groupby('region', as_index=False)['salary']
      .agg(['mean', 'median', 'count'])
      .sort_values('mean', ascending=False)
)
print("\n3. Average salary by region (geographic):")
print(salary_by_region)

# 4. Average salary by state (original)
salary_by_state = (
    df.groupby('state', as_index=False)['salary']
      .agg(['mean', 'median', 'count'])
      .sort_values('mean', ascending=False)
)
print("\n4. Top 10 states by average salary:")
print(salary_by_state.head(10))

# 5. Pivot table: level x role
salary_level_role_pivot = df.pivot_table(index='level', columns='role', values='salary', aggfunc='mean')
print("\n5. Salary matrix: level (rows) × role (columns):")
print(salary_level_role_pivot)

# 6. Pivot table: region x role
salary_region_role_pivot = df.pivot_table(index='region', columns='role', values='salary', aggfunc='mean')
print("\n6. Salary matrix: region (rows) × role (columns):")
print(salary_region_role_pivot)
```

## 8. Visualization 1: Average salary by role

This chart shows which role descriptors are paid more on average across all states in the dataset.


```python
plot_role = salary_by_role.sort_values('salary')

plt.figure(figsize=(9, 5))
plt.barh(plot_role['role'], plot_role['salary'])
plt.xlabel('Average annual salary (USD)')
plt.ylabel('Role descriptor')
plt.title('Average salary by role descriptor')

for i, v in enumerate(plot_role['salary']):
    plt.text(v, i, f' ${v:,.0f}', va='center')

plt.tight_layout()
plt.show()
```

## 9a. Visualization 2a: Salary heatmap by level and role

This shows how salaries vary across **career levels** (Junior, Mid-Level, Senior) and **role descriptors**, to identify where the highest-paying combinations are.



```python
# Level × Role heatmap
heatmap_level_role = salary_level_role_pivot.sort_index()

plt.figure(figsize=(10, 5))
im = plt.imshow(heatmap_level_role, aspect='auto', cmap='YlOrRd')
cbar = plt.colorbar(im, label='Annual salary (USD)')
plt.xticks(range(len(heatmap_level_role.columns)), heatmap_level_role.columns, rotation=45, ha='right')
plt.yticks(range(len(heatmap_level_role.index)), heatmap_level_role.index)
plt.title('Average salary by career level and role descriptor')
plt.ylabel('Career Level')
plt.tight_layout()
plt.show()

print("\nLevel × Role salary matrix:")
print(heatmap_level_role)
```
